## cleaning

Data tidiness issues





1.  Each variable should have its own column.


*   There are multiple columns (e.g., Yangon, Naypyitaw, Mandalay) representing branches as separate columns. These should be combined into a single Branch column with branch names as values.


2.  Each observation should have its own row.
*   nothing

3.  Each type of observational unit should form a table.
*   nothing



```python
branch_columns = ['Yangon', 'Naypyitaw', 'Mandalay']
Supermarket_data_copy = pd.melt(Supermarket_data_copy,
                                id_vars=['Invoice ID', 'Branch', 'Customer type', 'Gender', 'Product line', 'Unit price', 'Quantity', 'Tax 5%', 'Total', 'Date', 'Time', 'Payment', 'Rating'],
                                value_vars=branch_columns,
                                var_name='City')
Supermarket_data_copy = Supermarket_data_copy[Supermarket_data_copy['value'] == 1]
Supermarket_data_copy = Supermarket_data_copy.drop(columns=['value']).reset_index(drop=True)
```

```python
Supermarket_data_copy.sample(5,random_state=0)
```

### we can drop Branch column as it represents also city but with A,B,C as Yangon, Naypyitaw, Mandalay so they give the same information so it's better to drop 1 of them

```python
Supermarket_data_copy = Supermarket_data_copy.drop(columns=['Branch'])
```

```python
Supermarket_data_copy.sample(5)
```

## Data Quality issues
## 1. Missing Values:
* The Tax 5% column has 9 missing values.<br/>
* The Total column has 3 missing values.<br/>

## 2. Data Types Issues (Unit price, Total)

 * Unit price and Total should be float data type


## 3. Duplicate Records:

  * There are 6 duplicate rows in the dataset, which should be reviewed and potentially removed.

## 4. Inconsistent Data:

  * Some values in the Total column appear to be incorrectly formatted. For example, the value 3.405.255 has periods instead of commas or a decimal point.<br/>
  * Customer type  has 27 Nulls represented as dashes (-) records.
  * rating column have value equal 97

## 5. Invalid Data
 * Quantity column has negative values.<br/>

## 6. Inaccurate Data

  * Customer type has 'memberr' instead of 'member'.
  * The Time column has inconsistent formats (e.g., 8 - 30 PM), which could cause problems when performing time-based analyses.
  * Unit price column includes prices in USD .

# Handling Unidentified Customer Types

 In dealing with unidentified customer types, represented by (-), we have three options:

1. Create a New Type: Introduce a new category called "Unidentified Members" for these records.
2. Convert to Null Values: Replace these values with nulls and then either impute them or leave them as null.
3. Drop the Records: Remove these records from the dataset.

### Chosen Approach:

We have decided to drop these records. As they doesnt consume a large portion of the data. Imputing values for these records could potentially skew the analysis, so removing them will help maintain the accuracy and integrity of our results.

## code

```python
Supermarket_data_copy =Supermarket_data_copy[Supermarket_data_copy['Customer type'] != '-']
```
